### Visual 1

#### Histogram showing frequency of data recorded

```python
iris.plot.hist(subplots=True, layout=(2,2), figsize=(10, 10), bins=20)
```

### Visual 2

#### Kernel Density Plot of various species on the basis of sepal lenth

```python
g = sns.FacetGrid(iris, col='Classification')
g = g.map(sns.kdeplot, 'SL')
```

### Visual 2

#### Pair Plot

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris,hue="Classification",height=3, diag_kind="kde")
plt.show()
```
